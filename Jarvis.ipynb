```python
# Imports for recording
from io import BytesIO
from base64 import b64decode
from google.colab import output
from IPython.display import Javascript, Audio, display
import subprocess

# Imports for speaking
import speech_recognition as sr #pip install speechrecognition + PyAudio (manual downloading, coppy file name to install
from gtts import gTTS
from datetime import date, datetime #pip install datetime
import wikipedia, time
import librosa

RECORD = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=3):
  print("\nSpeak Now...")
  display(Javascript(RECORD))
  sec += 1
  s = output.eval_js('record(%d)' % (sec*1000))
  print("Done Recording!")
  b = b64decode(s.split(',')[1])
  with open('audio.webm','wb') as f:
    f.write(b)
  return 'audio.webm' # or return b
```

```python
def saying(text):
  tts = gTTS(text, lang='en') # AUS: com.au, UK: , tld='co.uk', CAN: ca
  tts.save('robot_saying.wav')
  robot_speak = 'robot_saying.wav' # save the name of the file that contains speaking audio

  return robot_speak
```

```python
# Start the chat
name = input("What is your name? ")
bot_name = input("What is your AI's name? ")
print("\nYou're chatting with %s... (Say 'Goodbye' to exit)\n%s: Hey %s, how are you?" %(bot_name, bot_name, name))

speaking = saying("Hey %s, how are you?" %name)
greetings = Audio(speaking, autoplay=True)
display(greetings)
duration1 = librosa.get_duration(filename='robot_saying.wav')
time.sleep(duration1 + 1.75)

# Start chatting here
for i in range(3):

    # Yui is listening...
    robot_ears = record(3) # Record for 5 sec
    subprocess.call(['ffmpeg', '-i', 'audio.webm', 'audio.wav']) # convert .webm file to .wav file
    sample_audio = sr.AudioFile('audio.wav')
    with sample_audio as audio_file:
        audio_content = sr.Recognizer().record(audio_file)
    try:
        you = sr.Recognizer().recognize_google(audio_content)
    except:
        you = "silent..."

    print("\nYou: " + you)
    #duration2 = librosa.get_duration(filename='audio.wav')
    robot_speak = Audio('audio.wav', autoplay=True)
    display(robot_speak)
    subprocess.call(['rm', 'audio.wav']) # removing audio file
    time.sleep(3 + 1.25)

    # Yui is thinking...
    robot_brain = ""
    if you == "silent...":
        robot_brain = "I can't hear you, try again"
    elif you == "hello":
        robot_brain = "Hello Duyen beautiful!"
    elif you == "who are you":
        robot_brain = "I'm your AI bot, created by Duyen."
    elif you == "today":
        today = date.today()
        robot_brain = today.strftime("%B %d, %Y")
    elif you == "time":
        now = datetime.now()
        robot_brain = now.strftime("%H hours %M minutes %S seconds")
    elif "who is" in you:
    	wiki = you[7:]
    	print(wiki)
    	try:
    		robot_brain = wikipedia.summary(wiki, sentences=1, auto_suggest=False)
    	except: 
    		robot_brain = "Page not found on Wikipedia"
    elif 'bye' in you:
        robot_brain = "Good bye, Duyen! Have a nice day!"
        print("Robot: " + robot_brain)
        robot_speak.say(robot_brain)
        robot_speak.runAndWait()
        break
    else:
        robot_brain = "Sorry, I have not been trained for this answer."

    # Yui is replying...
    #robot_brain = "Nice to see you." # Using tokenization to remove emojis and char from speech
    print("\nYui: " + robot_brain)

    speaking = saying(robot_brain)
    robot_speak = Audio(speaking, autoplay=True)
    display(robot_speak)
    duration3 = librosa.get_duration(filename='robot_saying.wav')
    time.sleep(duration3 + 1.5)
```
